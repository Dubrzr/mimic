```python
sig, fields = wfdb.rdsamp('mimic3wdb/matched/s00079/3887555_0020', pbdl=1, dldir='data/mimic3wdb-200315/s00079')

mwis, peaks_x, peaks_y, orig_x, orig_y, qrs_x, qrs_y, hrs_x, hrs_y = process(sig, fields)
```

```python
import matplotlib.pyplot as plt

def disp(a, b):
    fig, ax = plt.subplots(figsize=(20, 10))
    ax.plot(orig_x[a:b], orig_y[a:b], label='Original signal')
    ax.plot(qrs_x, qrs_y, 'ro', label='Rpeak')
    ax.set_xlim(a, b)
    plt.show()
    fig, ax_left = plt.subplots(figsize=(20, 10))
    ax_right = ax_left.twinx()
    #ax_left.plot(orig_x[a:b], orig_y[a:b], label='Original signal')
    ax_left.plot(mwis[+5+16+2+2+8:], label='MWI signal')
    ax_left.plot(qrs_x, qrs_y, 'ro', label='Rpeak')
    ax_left.plot(peaks_x, peaks_y, 'yo', label='Peaks')
    ax_right.plot(hrs_x[a:b], hrs_y[a:b], label='Heart rate', color='m')

    ax_left.set_title('QRS detector & Heart rate')

    ax_left.set_xlabel('Time (ms)')
    ax_left.set_ylabel('ECG V (mV)', color='b')
    ax_right.set_ylabel('Heart rate (bpm)', color='m')
    # Make the y-axis label, ticks and tick labels match the line color.
    ax_left.tick_params('y', colors='b')
    ax_right.tick_params('y', colors='m')
    # plt.savefig("test.png", dpi=600)
    ax_left.set_xlim(a, b)
    plt.show()

# for i in range(100):
#     a = 0 + i * 10000
#     b = a + 10000
#     disp(a,b)
disp(79000,79500)
```
